```python
import pandas as pd
import numpy as np
xls = pd.ExcelFile('/home/techverx/Downloads/Data.xlsx')
df1 = pd.read_excel(xls, 'Sample Data',header=2)
df1=df1.drop(["Unnamed: 0","Unnamed: 1"],axis=1)

month_index=df1["Month"].unique()
year_index=df1["Year"].unique()
m=df1["WarehouseName"].unique()
gf={"warehousename":m}
fd=pd.DataFrame(gf)
fd=fd.set_index("warehousename")

```

```python
h=[]
for i in range(len(year_index)):
    for j in range(len(month_index)):
        newdf=df1[(df1.Month==month_index[j])&(df1.Year==year_index[i])]
        h.append(newdf)

h = [l for l in h if not l.empty]
```

```python
j=[]
for i in h:
    newdf1=i.rename(columns={'Gross Sales (Qty)':'Gross Sales (Qty)'+i["Year"].values[0].astype('str')+i["Month"].values[0], 'Qty - Net Sales': 'Qty - Net Sales'+i["Year"].values[0].astype('str')+i["Month"].values[0],'Gross Sales': 'Gross Sales'+i["Year"].values[0].astype('str')+i["Month"].values[0],'Value - Net Sales':'Value - Net Sales'+i["Year"].values[0].astype('str')+i["Month"].values[0],'COGS':'COGS'+i["Year"].values[0].astype('str')+i["Month"].values[0],'Gross Profit':'Gross Profit'+i["Year"].values[0].astype('str')+i["Month"].values[0],'Margin':'Margin'+i["Year"].values[0].astype('str')+i["Month"].values[0],'Discount Amount':'Discount Amount'+i["Year"].values[0].astype('str')+i["Month"].values[0],'Discount%':'Discount%'+i["Year"].values[0].astype('str')+i["Month"].values[0],'Tax Amount':'Tax Amount'+i["Year"].values[0].astype('str')+i["Month"].values[0] })
#     newdf1.set_index("WarehouseName")
    newdf1.reset_index(drop=True,inplace=True,)
    j.append(newdf1.set_index("WarehouseName").sort_index())
j.insert(0,fd) 
k=pd.concat(j,axis=1)
k=k.drop(['Year','Month'],axis=1)
k=k.fillna('')
k
```
