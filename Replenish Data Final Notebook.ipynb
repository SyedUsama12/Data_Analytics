```python
import pandas as pd
import numpy as np
xls = pd.ExcelFile('/home/techverx/Downloads/Replenishment Data Working V1 - 02-11-2021.xlsx')
df1 = pd.read_excel(xls, 'Required Data',header=3)
df2 = pd.read_excel(xls, 'Data', header=1)
d=df2.drop(["Year","Month","Unnamed: 15","Unnamed: 0","Date"],axis=1)
# d.sort_values(by=['ProductID','SizeID'],inplace=True)
d['SizeID']=d['SizeID'].astype('str')
```

```python
table = pd.pivot_table(data=d, index=['ProductID','ColorName','SizeID',"Shop"],
                        values=['Opening Stk','Sales (Qty)','Transfer IN','Transfer Out','Closing Stk','Transit Qty','Total Qty'],
                        aggfunc='sum'
                        ).sort_index()
```

```python
df = table.groupby(level=[0,1,2]).sum()
df.index = pd.MultiIndex.from_arrays([df.index.get_level_values(0), 
                                       df.index.get_level_values(1),
                                       df.index.get_level_values(2).astype("str")+ '_sum', 
                                       len(df.index) * [''],
                                      len(df.index) * [''],
                                      len(df.index) * ['']
                                     ])

      
```

```python
#groupby and create sum dataframe by levels 0,1
df1 = table.groupby(level=[0,1]).sum()
df1.index = pd.MultiIndex.from_arrays([df1.index.get_level_values(0), 
                                       df1.index.get_level_values(1)+ '_sum', 
                                       len(df1.index) * [''],
                                       len(df1.index) * [''],
                                       len(df1.index) * ['']
                                       
                                      
                                      ])

                  



```

```python
df2 = table.groupby(level=0).sum()
df2.index = pd.MultiIndex.from_arrays([df2.index.values + '_sum',
                                       len(df2.index) * [''] ,
                                       len(df2.index) * [''],
                                       len(df2.index) * [''],
                                       len(df2.index) * ['']
                                       
                                       
                                       
                                       
                                       
                                       ])
```

```python
done = pd.concat([table,df ,df1, df2]).sort_index(level=[0])
```
